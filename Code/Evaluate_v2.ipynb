#### CLIP-Score

```python
import torch
from transformers import CLIPProcessor, CLIPModel
from PIL import Image

# CLIP 모델 및 프로세서 로드
device = "cuda" if torch.cuda.is_available() else "cpu"
model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32").to(device)
processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")

def compute_clip_score(images, texts):
    scores = []
    for image_path, text in zip(images, texts):
        # 이미지 로드 및 전처리
        image = Image.open(image_path).convert("RGB")
        inputs = processor(text=[text], images=image, return_tensors="pt", padding=True).to(device)

        # CLIP 임베딩 추출
        outputs = model(**inputs)
        image_features = outputs.image_embeds
        text_features = outputs.text_embeds

        # 코사인 유사도 계산
        cos_sim = torch.nn.functional.cosine_similarity(image_features, text_features)
        scores.append(cos_sim.item())
    return scores

# 원본 데이터와 생성된 데이터 비교
original_clip_scores = compute_clip_score(original_images, original_texts)
dalle_clip_scores = compute_clip_score(original_images, dalle_texts)
imagen_clip_scores = compute_clip_score(imagen_images, imagen_texts)
sd_clip_scores = compute_clip_score(sd_images, sd_texts)
figma_clip_scores = compute_clip_score(figma_images, figma_texts)
```

```python
import torch
import numpy as np
from torchvision.models import inception_v3
from torchvision.models.feature_extraction import create_feature_extractor
from torchvision import transforms
from PIL import Image
from scipy.spatial.distance import cdist

# Inception 모델 로드 및 중간 레이어 출력 설정
device = "cuda" if torch.cuda.is_available() else "cpu"
inception = inception_v3(pretrained=True).to(device).eval()
return_nodes = {'avgpool': 'features'}  # avgpool 레이어 출력 사용
inception_extractor = create_feature_extractor(inception, return_nodes=return_nodes).to(device).eval()

# 이미지 전처리 함수
def preprocess_images(image_paths):
    transform = transforms.Compose([
        transforms.Resize((299, 299)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ])
    features = []
    for image_path in image_paths:
        try:
            image = Image.open(image_path).convert("RGB")
            input_tensor = transform(image).unsqueeze(0).to(device)

            # 특징 벡터 추출
            with torch.no_grad():
                feature = inception_extractor(input_tensor)['features'].cpu().numpy().flatten()
            features.append(feature)
        except Exception as e:
            print(f"Error processing image {image_path}: {e}")
    return np.array(features)

# KID 계산 함수 (RBF 커널 사용)
def compute_kid(features1, features2):
    sigma = np.median(cdist(features1, features2))  # 거리의 중간값 사용
    kernel = lambda x, y: np.exp(-cdist(x, y, metric="euclidean")**2 / (2 * sigma**2))
    k11 = kernel(features1, features1)
    k22 = kernel(features2, features2)
    k12 = kernel(features1, features2)
    return k11.mean() + k22.mean() - 2 * k12.mean()

# 이미지 경로 설정
original_images = [
    "/content/drive/MyDrive/[빅컴]프로젝트/vlm-main/Generate_Image_Data/Label_Image/11739.jpg",
    "/content/drive/MyDrive/[빅컴]프로젝트/vlm-main/Generate_Image_Data/Label_Image/60871.jpg",
    "/content/drive/MyDrive/[빅컴]프로젝트/vlm-main/Generate_Image_Data/Label_Image/1093531.jpg",
    "/content/drive/MyDrive/[빅컴]프로젝트/vlm-main/Generate_Image_Data/Label_Image/1153777.jpg",
    "/content/drive/MyDrive/[빅컴]프로젝트/vlm-main/Generate_Image_Data/Label_Image/1185276.jpg"
]

dalle_images = [
    "/content/drive/MyDrive/[빅컴]프로젝트/vlm-main/Generate_Image_Data/Generate_Image/dalle_11739.jpg",
    "/content/drive/MyDrive/[빅컴]프로젝트/vlm-main/Generate_Image_Data/Generate_Image/dalle_60871.jpg",
    "/content/drive/MyDrive/[빅컴]프로젝트/vlm-main/Generate_Image_Data/Generate_Image/dalle_1093531.jpg",
    "/content/drive/MyDrive/[빅컴]프로젝트/vlm-main/Generate_Image_Data/Generate_Image/dalle_1153777.jpg",
    "/content/drive/MyDrive/[빅컴]프로젝트/vlm-main/Generate_Image_Data/Generate_Image/dalle_1185276.jpg"
]

imagen_images = [
    "/content/drive/MyDrive/[빅컴]프로젝트/vlm-main/Generate_Image_Data/Generate_Image/imagen3_11739.jpg",
    "/content/drive/MyDrive/[빅컴]프로젝트/vlm-main/Generate_Image_Data/Generate_Image/imagen3_60871.jpg",
    "/content/drive/MyDrive/[빅컴]프로젝트/vlm-main/Generate_Image_Data/Generate_Image/imagen3_1093531.jpg",
    "/content/drive/MyDrive/[빅컴]프로젝트/vlm-main/Generate_Image_Data/Generate_Image/imagen_1153777.jpg",
    "/content/drive/MyDrive/[빅컴]프로젝트/vlm-main/Generate_Image_Data/Generate_Image/imagen_1185276.jpg"
]

sd_images = [
    "/content/drive/MyDrive/[빅컴]프로젝트/vlm-main/Generate_Image_Data/Generate_Image/stable_diffusion_11739.png",
    "/content/drive/MyDrive/[빅컴]프로젝트/vlm-main/Generate_Image_Data/Generate_Image/stable_diffusion_60871.png",
    "/content/drive/MyDrive/[빅컴]프로젝트/vlm-main/Generate_Image_Data/Generate_Image/stable_diffusion_1093531.png",
    "/content/drive/MyDrive/[빅컴]프로젝트/vlm-main/Generate_Image_Data/Generate_Image/stable_diffusion_1153777.png",
    "/content/drive/MyDrive/[빅컴]프로젝트/vlm-main/Generate_Image_Data/Generate_Image/stable_diffusion_1185276.png"
]

figma_images = [
    "/content/drive/MyDrive/[빅컴]프로젝트/vlm-main/Generate_Image_Data/Generate_Image/FIGMA_11739.png",
    "/content/drive/MyDrive/[빅컴]프로젝트/vlm-main/Generate_Image_Data/Generate_Image/FIGMA_60871.png",
    "/content/drive/MyDrive/[빅컴]프로젝트/vlm-main/Generate_Image_Data/Generate_Image/FIGMA_1093531.png",
    "/content/drive/MyDrive/[빅컴]프로젝트/vlm-main/Generate_Image_Data/Generate_Image/FIGMA_1153777.png",
    "/content/drive/MyDrive/[빅컴]프로젝트/vlm-main/Generate_Image_Data/Generate_Image/FIGMA_1185276.png"
]

# 특징 추출
original_features = preprocess_images(original_images)
dalle_features = preprocess_images(dalle_images)
imagen_features = preprocess_images(imagen_images)
sd_features = preprocess_images(sd_images)
figma_features = preprocess_images(figma_images)

# KID 점수 계산 및 출력
print("KID Score (DALL-E):", compute_kid(original_features, dalle_features))
print("KID Score (Imagen):", compute_kid(original_features, imagen_features))
print("KID Score (Stable Diffusion):", compute_kid(original_features, sd_features))
print("KID Score (Figma):", compute_kid(original_features, figma_features))
```
